```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad
from base64 import b64encode
from google.colab import drive
import os
from Crypto.Util.Padding import unpad
from base64 import b64decode
```

```python
def symmetric_encryptor(input_message:str , key:str):
    # Write a program to encrypt the given string input
    # CODE HERE:

    # Convert key to bytes
    key_bytes = key
    # key_bytes = key.encode()

    # Initialize the cipher object with the key and AES mode
    cipher = AES.new(key_bytes, AES.MODE_ECB)

    # Pad the input message to multiples of 16 bytes (AES block size)
    padded_message = pad(input_message.encode(), AES.block_size)

    # Encrypt the padded message
    encrypted_message = cipher.encrypt(padded_message)

    # Encode the encrypted message in base64 for readability
    cipher_text = b64encode(encrypted_message).decode()

    return cipher_text

# message = "Hello, world!"
# key = "ThisIsASecretKey"
# encrypted_message = symmetric_encryptor(message, key)
# print("Encrypted Message:", encrypted_message)
```

```python
def symmetric_decryptor(input_ciphertext:str , key:str):
    # Write a program to decrypt the given string input
    # CODE HERE:

    # Convert key to bytes
    key_bytes = key
    # key_bytes = key.encode()

    # Initialize the cipher object with the key and AES mode
    cipher = AES.new(key_bytes, AES.MODE_ECB)

    # Decode the base64 encrypted message
    encrypted_message = b64decode(input_ciphertext)

    # Decrypt the encrypted message
    decrypted_message = cipher.decrypt(encrypted_message)

    # Unpad the decrypted message
    unpadded_message = unpad(decrypted_message, AES.block_size)

    # Decode the plaintext message
    plain_text = unpadded_message.decode()

    return plain_text
```

```python
def symmetric_verifier(plaintext:str, ciphertext:str, key:str):
    # Verify the correctness of the decryption process by
    # comparing the original message with the decrypted message.
    # output should be boolean(ture or false)
    # CODE HERE:

    result = symmetric_decryptor(ciphertext , key)
    if result == plaintext:
        return True
    else:
        return False

```

## Part 3: Digital signature

Write a function to sign a given message in the text file (Confidentail message) using private key which was made in the part 2.   
You have to sign the hash of the message, not the whole message (why?). (Use SHA-256 for hashing)

```python
from Crypto.Hash import SHA256
from Crypto.Signature import pkcs1_15
```

```python
def sign_message(private_key:str, input_message: str):
    # Write a code to sign a message using your private key.
    # CODE HERE:

    # Import the private key
    key = RSA.import_key(private_key)

    # Calculate the hash of the input message
    hash_message = SHA256.new(input_message.encode())

    # Create a signer object
    signer = pkcs1_15.new(key)

    # Sign the hash of the message
    signature = signer.sign(hash_message)
    message_signature = signature.hex()

    return message_signature
```
